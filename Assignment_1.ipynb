# Performing Visualizations

```python
#First we will create an histogram for Petal length for each classification using seaborn
sns.histplot(data=iris_data, x="PL", hue="Classification")
plt.title("Plot for Petal Lenght for each classification")
plt.xlabel("Petal Length", fontsize=15)
```

From the above histogram it is clear that iris_virginica has the larget PL and Iris_setosa has lowest PL.Also we can see that iris_versicolor and iris_setosa has similar petal length.

```python
#Swarmplot for identifying the sepal length for each classification
sns.swarmplot(x=iris_data['Classification'], y=iris_data['SL'])
plt.title("Swarmplot for Petal length for each classification")
plt.xlabel("Classification", fontsize=15)
plt.ylabel("SepalLength", fontsize=15)
```

From this swarmplot we can clearly see the difference in Sepal length in all the three classifications.
Iris-setosa has lowest sepal length and Iris-virginical has largest sepal length

```python
#Regplot for Petal width and Sepal width
sns.lmplot(data=iris_data, x="SW", y="PW", hue="Classification")
plt.title("Plot for Sepal width and Petal width")
plt.xlabel("Sepal width", fontsize=15)
plt.ylabel("Petal width", fontsize=15)
```

Positive correlation for SW & PW for all the three classification.
